```python
import pandas as pd 
import numpy as np
import matplotlib as pyplt
import seaborn
```

```python
#import from Data
```

```python
df= pd.read_csv("cleaning data.csv")
df
```

```python
#Check for null value
```

```python
df.isnull().sum()
```

```python
s=df.drop('Salary',axis=1)
s
```

```python
y=df['Salary']
y
```

```python
#One hot encoding 
```

```python
G=pd.get_dummies(df['Gender'],drop_first=True)
G
```

```python
E=pd.get_dummies(df['Education Level'],drop_first=True)
E
```

```python
J=pd.get_dummies(df['Job Title'],drop_first=True)
J
```

```python
x = s.drop(['Job Title','Education Level','Gender'],axis=1)
x
```

```python
x=pd.concat([x,G,E,J],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x, y.values)
predict=reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
error=np.sqrt(mean_squared_error(y,predict))
error
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(random_state=0)
clf.fit(x,y.values)
```

```python
prediction=clf.predict(x)
```

```python
error=np.sqrt(mean_squared_error(y,prediction))
error=("{:,.02f}".format(error))
error
```

```python
from sklearn.ensemble import RandomForestClassifier
rclf=RandomForestClassifier(random_state=0)
rclf.fit(x,y.values)
```

```python
pre=rclf.predict(x)
```

```python
error=np.sqrt(mean_squared_error(y,pre))
error=("{:,.02f}".format(error))
error
```
